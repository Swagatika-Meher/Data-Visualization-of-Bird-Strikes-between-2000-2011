```python
# Data Cleaning, Data Pre-processing, Data Analysis

import pandas as pd
import numpy as np
bird_strikes = pd.read_csv("/content/Bird Strikes.csv")      #copy the file path
bird_strikes.head()                                               #Reading Dataset
```

```python
# All Columns

columns = list(bird_strikes)
```

**Data Analysis**

```python
# Converting date format
bird_strikes['FlightDate'] = pd.to_datetime(bird_strikes['FlightDate'])
bird_strikes.head()
```

```python
#Creating Year and Month columns
bird_strikes['Year'] = bird_strikes['FlightDate'].dt.year
bird_strikes['Month'] = bird_strikes['FlightDate'].dt.month
bird_strikes.head()
```

```python
# Creating DataFrame only with neccessary values.
import matplotlib.pyplot as plt
import seaborn as sns
bird_strikes02 = bird_strikes[['Record ID','Wildlife: Number Struck Actual', 'Remains of wildlife collected?', "Remains of wildlife sent to Smithsonian", "Number of people injured"]]
# Checking the correlation
plt.figure(figsize=(10,5))
sns.heatmap(bird_strikes02.corr(method='pearson'), annot=True, vmin=-1, vmax=1, cmap='YlGnBu')
```
